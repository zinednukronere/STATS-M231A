```python
from transformers import pipeline
unmasker = pipeline('fill-mask', model='bert-base-uncased')
unmasker("Hello I'm a [MASK] model.")
```

```python
unmasker("I went to the [MASK] to see a movie.")
```

```python
unmasker("I drank some [MASK] to wake up.")
```

```python
unmasker("I did not [MASK] last night.")
```

```python
germanUnmasker = pipeline('fill-mask', model='bert-base-german-cased')
```

```python
germanUnmasker("ich mag meine [MASK] fahren.")
```

```python
germanUnmasker("ich mag [MASK] essen am Frühstück.")
```

```python
germanUnmasker("Ich gehe gerne ins [MASK].")
```
