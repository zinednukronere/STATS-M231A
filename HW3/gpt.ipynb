```python
from transformers import pipeline, set_seed
generator = pipeline('text-generation', model='gpt2')
```

```python
set_seed(42)
generator("Hello, I'm a language model,", max_length=30, num_return_sequences=5)
```

```python
generator("The meaning of life is", max_length=30, num_return_sequences=5)
```

```python
generator("The key to happiness", max_length=30, num_return_sequences=5)
```

```python
generator("I need to work but", max_length=30, num_return_sequences=5)
```

```python
generator("The man was sitting alone in a cafe", max_length=300, num_return_sequences=1)
```

```python
story_gen = pipeline("text-generation", "pranavpsv/gpt2-genre-story-generator")
```

```python
story_gen("<BOS> <drama> The man  was sitting alone in a cafe",max_length=300, num_return_sequences=1)
```
